```python
import os
import sys
import pandas as pd
from bs4 import BeautifulSoup
   
folder = os.listdir('Routes')
folder.sort()
folder.pop(0)
folder.pop(0)

j=0
for s in folder:
    i = s.index('.')
    j = j+1
    ans = s[0:i]
    ans = ans+".csv"
    print(ans , end=" ")
    print(j)

    data = []
    path = 'Routes/'
    path = path +s
    # for getting the header from
    # the HTML file
    soup = BeautifulSoup(open(path),'html.parser')
    header = soup.find_all("table")
    # print(len(header))
    header2 = header[1].find_all("tr")
    for i in header2:
        temp = i.find_all("td")
        l = []
        for s in temp:
            k= " ".join((s.get_text()).split())
            l.append(k)
        data.append(l)
    data.pop(1)
    data.pop(0)

    for i in data:
        i.pop(0)

    dataFrame = pd.DataFrame(data = data)

    dataFrame.to_csv(ans)
    
```
